# Video 6 Pandas: Creacion y borrado de columnas en un Dataframe

```python
# Si quisieramos facilmente crear UNA NUEVA COLUMNA asignando todos los valores en cero
oscars_short = pd.read_csv('Oscars-demographics-DFE.csv', encoding ='latin-1',\
                          usecols=['birthplace','date_of_birth','race_ethnicity','religion','sexual_orientation','year_of_award',\
                                  'award','movie','person'])
oscars_short.tail()
```

```python
oscars_short
```

```python
# verifico haciendo un values_count()
oscars_short['last_ten_years'].value_counts()
```

```python
# Si quiero eliminar la columna que creamos, solo hacemos un drop
oscars_short.drop(columns='last_ten_years', inplace=True)
```

```python
oscars_short
```

```python
# Si quisieramos agregar la columna en un lugar especifico del data frame (por ejemplo entre la fecha de nacimeint oy la raza del grupo etnico)

# 1) calculamos la posicion con get_loc()
date_of_birth_position=oscars_short.columns.get_loc('date_of_birth')+1
date_of_birth_position
```

```python
# ahora que tenemos la posicion de la nueva columna, hacemos un insert y le vamos a indicar que la posicion es la que calculamos (date of birth +1) y le ponemos el nombre
#y le indicamos el valor que puede tomar

oscars_short.insert(loc=date_of_birth_position, column='last_ten_years', value=(oscars_short['year_of_award']>2009))
oscars_short
```
